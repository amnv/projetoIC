### Pegando dados do Google Custom search

Api de pesquisa google https://developers.google.com/custom-search/

Versão gratuita permite apenas 10 requisições por dia retornando no máximo 10 resultados por consulta

```python
import requests
import datetime
import nltk
```

```python
text = []
next_page = 1
apis = open("../data/keys/google_api_key", "r")
free_api = apis.readline()
paid_api = apis.readline()

while True:
    url = "https://www.googleapis.com/customsearch/v1?key={0}&cx=012152041422378146968:kcnelx4z5qg&q=Tissue source such as&start={1}&exactTerms=Tissue source such as".format(paid_api, next_page)    
    
    # Download data
    r = requests.get(url)

    # Parsing to Json
    data = r.json()
    if int(data["searchInformation"]["totalResults"]) > 0:
        # Get relevant information
        for i in data["items"]:
            print(i["snippet"] + "\n")
            text.append(i["snippet"] + "\n")

        # Find next item
        next_page += data["queries"]["nextPage"][0]["startIndex"]
        print(next_page)
    else: 
        break

# Write result to a file
now = datetime.datetime.now()
file_name = '../data/extractions/extraction_{0}.txt'.format(now.date())
with open(file_name, 'w') as f:
    for item in text:
        f.write(item)    
```

```python
now = datetime.datetime.now()
file_name = '../data/google_extractions/extraction_{0}_nao_exato.txt'.format(now.date())
with open(file_name, 'w') as f:
    for item in text:
        f.write(item) 
```

```python
import re

relevant_words = []
for i in text:
    count = 0
    previus_count = 0
    x = re.split("tissue source.*such as.", i)
    if len(x) > 1:
        print("sentença: " + x[-1])
        tokens = nltk.word_tokenize(x[-1])
        pos_tag = nltk.pos_tag(tokens)

        for i, j in pos_tag:
            print(i + " " + j)
            if j.startswith("N") | j.startswith("J"):
                print("palavras relevantes: " + i)
                
                if count - 1 == previus_count:
                    previus_count = count
                    relevant_words.append(" " + i)
                else:
                    relevant_words.append("\n" + i)
            count += 1
        print("\n")
        relevant_words.append("\n")
        
file_name = '../data/google_extractions/relevant_words_{0}.txt'.format(now.date())
with open(file_name, 'w') as f:
    for item in relevant_words:
        f.write(item)  
```
