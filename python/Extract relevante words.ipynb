```python
import nltk
from nltk.corpus import treebank
import re
import glob
from datetime import datetime

now = datetime.now()
```

```python
files_name = glob.glob('../data/extractions/web_search/extraction_*.txt')

for file in files_name:
    print(file)
    with open(file, 'r') as f:
        count = 0
        count_such = 0
        for i in f:
            if "tissue source" in i:
                #print(i)
                count_such += 1
            count += 1
        print("total: " + str(count))
        print("such sentences: " + str(count_such))
```

```python
# Extraindo tissue source 
data = "2019-04-07"
text = open("../data/extractions/web_search/extraction_{0}.txt".format(data), "r")
relevant_words = []
for i in text:
    count = 0
    previus_count = 0
    x = re.split("tissue source.*such as.", i)
    if len(x) > 1:
        print("sentença: " + x[-1])
        tokens = nltk.word_tokenize(x[-1])
        pos_tag = nltk.pos_tag(tokens)

        for word, tag in pos_tag:
            print(word + " " + tag)
            if not tag.startswith("V") and not word.startswith("."):
                print("palavras relevantes: " + word)
                if count - 1 == previus_count:
                    previus_count = count
                    relevant_words.append(" " + word)
                else:                   
                    relevant_words.append("\n" + word)
            else: break
                
            count += 1
        print("\n")
        relevant_words.append("\n")
        
file_name = '../data/extractions/relevant_words/relevant_words_{0}.txt'.format(data)
with open(file_name, 'w') as f:
    for item in relevant_words:
        f.write(item)  
```
