```python
import pronto
ont = pronto.Ontology('../cl.owl')
```

## Cell type extractions

```python
count = 0
for term in ont:
    print(term)
    count += 1
print(count)
```

```python
cells_type = []
for term in ont:
    if term.id.startswith("CL"):
        cells_type.append(term.name)
```

### Quantidade de tipos de célula

```python
len(cells_type)
```

```python
# Tipo de célula e descrição
cell_desc = []
for term in ont:
    if term.id.startswith("CL"):
        cell_desc.append("term: " + term.name)
        cell_desc.append("desc: " + term.desc + "\n")
        #print(term.id)
```

```python
cells_type
```

```python
# Salvando dados
fp = open("../data/extractions/cell_type.txt", "w")
for i in cells_type:
    fp.write(i + "\n")

fp.close()
```

```python
fp = open("../data/extractions/cell_desc.txt", "w")
for i in cell_desc:
    fp.write(i + "\n")

fp.close()
```

## Tissue source extractions

```python
ont = pronto.Ontology('../ktao-merged.owl')
```

```python
print(ont)
```

```python
tissue_source_UBERON = []
tissue_source = []
for term in ont:
    if "bone marrow" in term.name or "adipose tissue" in term.name or "umbilical cord" in term.name:
        print(term)
```

```python
tissue_source_UBERON = []
tissue_source_UBERON_desc = []
tissue_source = []
tissue_source_desc = []
for term in ont:
    if term.id.startswith("UBERON"):
        tissue_source_UBERON.append(term.name)
        tissue_source_UBERON_desc.append("term: " + term.name)
        tissue_source_UBERON_desc.append("desc: " + term.desc)
    tissue_source.append(term.name)
    tissue_source_desc.append("term: " + term.name)
    tissue_source_desc.append("desc: " + term.desc)
```

```python
def writer(file_name, data):
    fp = open(file_name, "w")#"../data/extractions/cell_desc.txt", "w")
    for i in data:
        fp.write(i + "\n")

    fp.close()
```

```python
writer("../data/extractions/tissue_source_UBERON.txt", tissue_source_UBERON)
writer("../data/extractions/tissue_source_UBERON_desc.txt", tissue_source_UBERON_desc)
writer("../data/extractions/tissue_source.txt", tissue_source)
writer("../data/extractions/tissue_source_desc.txt", tissue_source_desc)
```
